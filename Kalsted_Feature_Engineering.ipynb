# Feature Engineering Part One

```python
#Here we are creating a separate object called "label" using the data from the median house value column

label = df.median_house_value.copy()
```

```python
#Here is the data

label
```

```python
#Now we are creating a separate dataframe called "features" that has all the original data except median house value

features = df.drop(columns = ['median_house_value'])
```

```python
#Here isa brief look at features

features.info()
```

```python
#Here we are selecting all float datatypes from features

features.select_dtypes('float')
```

```python
#Here we are importing the stats library

import scipy.stats as stats
```

```python
#Here we are selecting float datatypes from the features dataframe
#We are applying a z-score on every column
#Simply put, a z-score (also called a standard score) gives you an idea of how far from the mean a data point is. 
#But more technically it’s a measure of how many standard deviations below or above the population mean a raw score is.

feat1 = features.select_dtypes('float').apply(lambda x: stats.zscore(x))
```

```python
#Here are all columns expressed in z-scores

feat1
```

```python
#Here we are making sure that all of our floats are expressed with two decimal points

pd.options.display.float_format = '{:.2f}'.format
```

```python
#Here we are getting the mean and standard deviation from each column
#As you can see the mean is 0 and std is 1. This is the definition of standardization
#We standardized our data with the z-score earlier

feat1.agg(['mean', 'std'])
```

# Feature Engineering Part Two

```python
#Here we are converting the categorical values in ocean_proximity to dummy variables

dummies = pd.get_dummies(features.ocean_proximity)
```

```python
#Here we are combining features, feat1, dummies, and df,income_cat all together

features = pd.concat([feat1, dummies, df.income_cat], axis = 1)
```

# Splitting Data Into Training and Test Sets

```python
#here we are taking a quick look at the features dataframe 

features
```

```python
#Here we are making our test size 20%
#It is common to use 80% of the data for training and 20% for the test set

test_size = 0.2
```

```python
#Here we are drawing a random sample using the .sample method
#sample() is an inbuilt function of random module in Python that returns a particular length list of items chosen 
#from the sequence i.e. list, tuple, string or set. Used for random sampling without replacement.
#frac = test_size meaning we are getting a random sample of 20% of the data
#You can pass any number into random_state. As long as you keep the number the same, we will get the same sample
#That allows for reproducibility 

X_test = features.sample(frac = test_size, random_state = 123)
```

```python
#Here we are taking a look at the random sample we drew for our X_test
#4087 rows is 20% of 20,433 rows we have in features
#We must be careful of sampling bias. Does the sample represent the whole?

X_test
```

```python
#Here we are checking the frequencies of the data in our test set
#We do this with "normalize"
#So we have 25%, 25%, 25%, 20%, and 5%

X_test.income_cat.value_counts(normalize = True)
```

```python
#Here we are looking at the frequencies in the main dataframe
#Looks like the random sample has the same distribution as the main dataset
#This is a representative sample

features.income_cat.value_counts(normalize = True)
```

```python
#Here we are looking at the row labels for the test set

X_test.index
```

```python
#Here we are creating our training set
#We are saying any row NOT in X_test will be in X_train
#This is done with the tilde ~ symbol

X_train = features.loc[~features.index.isin(X_test.index)].copy()
```

```python
#Here is a quick look at X_train
#16,346 rows is 80% of 20,433 rows in the original dataframe

X_train
```

```python
#And here we are once again checking the distribution
#The distribution matches X_test and the original dataframe
#We can compare apples to apples now

X_train.income_cat.value_counts(normalize = True)
```

```python
#Here we are shuffling around the training set
#Frac = 1 because we are using the whole X_train, just shuffling it around

X_train = X_train.sample(frac = 1, random_state = 123)
```

```python
#Here is X_train. As you can see the rows are not in order since we shuffled them around

X_train
```

```python
#Here we are dropping income_cat because for now we are only using numerical data for our model

X_train.drop(columns = ['income_cat'], inplace = True)
```

```python
#Here we are dropping income_cat because for now we are only using numerical data for our model

X_test.drop(columns = ['income_cat'], inplace = True)
```

```python
#We already created our X values
#Now we are creating our y values

y_train = label.loc[X_train.index]
y_test = label.loc[X_test.index]
```

```python
#Here is a quick look at y_train

y_train
```

# Training the Machine Learning Model Using Random Forest

```python
#Here we are importing the Random Forest Regressor algorithm from the sklearn ensemble module
#The sklearn.ensemble module includes two averaging algorithms based on randomized decision trees: 
#the RandomForest algorithm and the Extra-Trees method.
#The RandomForestRegressor can capture non-linear relationships

from sklearn.ensemble import RandomForestRegressor
```

```python
#Here we are creating the regressor itself

forest_reg = RandomForestRegressor(random_state = 42, n_estimators = 500, max_features = 'sqrt', 
                                   max_depth = 75, min_samples_split = 2)
```

```python
#Here we are fitting the training data to the regressor

forest_reg.fit(X_train, y_train)
```

```python
#Before we fit our test data we can score the training data
#We are seeing if there's overfitting (1) or underfitting (0)
#We have a pretty high value in .98. This could be a sign of overfitting

forest_reg.score(X_train, y_train)
```

```python
#Here we are importing mean squared error
#The closer to 0 MSE is, the better the fit of our regression line is

from sklearn.metrics import mean_squared_error
```

```python
#Here we are making predictions on the training set

pred = forest_reg.predict(X_train)
```

```python
#Here are the y-predictions based on X_train

pred
```

```python
#Here we are caluclating mean squared error and root mean squared error

forest_mse = mean_squared_error(y_train, pred)
forest_rmse = np.sqrt(forest_mse)
```

```python
print(forest_mse)
print(forest_rmse)
```

# Evaluating the Model on the Test Set

```python
#Now we are scoring the test set
#We originally scored the training set and got a great value of .98
#Here we have .83, much less than .98. Definitely a sign of overfitting, but .83 is still a good number

forest_reg.score(X_test, y_test)
```

```python
#Here we are making predictions on the test set

pred = forest_reg.predict(X_test)
```

```python
#Here are the y-predictions based on X_test

pred
```

```python
#Here we are caluclating mean squared error and root mean squared error

forest_mse = mean_squared_error(y_test, pred)
forest_rmse = np.sqrt(forest_mse)
```

```python
print(forest_mse)
print(forest_rmse)
```

```python
#Here we are creating a dataframe that will compare the predicted values to the true values

comp = pd.DataFrame(data = {"True_V": y_test, "Pred": pred})
```

```python
#Here is what that dataframe looks like

comp
```

```python
#Here we are calculating the difference between the true and predicted values
#We are making sure the difference is expressed in absolute value terms

ae = comp.True_V.sub(comp.Pred).abs()
```

```python
#Here are the calculated differences

ae
```

```python
#Here we are calculating mean absolute error
#Which is the average of the absolute error

mae = ae.mean()
```

```python
#Here is the calculated mean absolute error

mae
```

# Feature Importance

```python
#Feature importances are provided by the fitted attribute feature_importances_ 
#and they are computed as the mean and standard deviation of accumulation of the impurity decrease within each tree.

forest_reg.feature_importances_
```

```python
#Here we are creating a series from the numpy array of feature importances

feature_imp = pd.Series(data = forest_reg.feature_importances_, index = X_train.columns).sort_values(ascending = False)
```
